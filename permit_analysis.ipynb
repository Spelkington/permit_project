```python
# How many months did the number of SLC residential permits exceed 100?
```

```python
# List out the dataframe columns that we have to work with.
list(df)
```

```python
# Create a frame of rows Salt Lake City entries with permits greater than 100.
slc_permit_df = df[
    (df['City'] == 'Salt Lake City') & 
    (df['New Dwelling Units'] > 100)
]
```

```python
# What was the total value of non-residential construction permits,
# including both new and additional permits, in Provo during May of 2014
```

```python
# Retrieve the Provo, May 2014 row from the frame.
provo_df = df[
    (df['City'] == 'Provo') &
    (df['Month'] == 'May') &
    (df['Year'] == 2014)
]
```

```python
provo_df
```

```python
# Sum the two non-residential valuations
non_res_sum = provo_df['New NonResidential Valuation'] + provo_df['Value of Nonresidential Additions']
```

```python
# Multiply this sum by 1000, to signify
non_res_sum * 1000
```

```python
del(provo_df)
```

```python
# In how many months since 2000 did the value of
# Moab residential construction permits exceed 1 million?
```

```python
# Create a new dataframe containing only the rows meeting the criteria,
# where the location is Moab, the months are only those from 2000 and
# beyond, and the sum of residential construction values are > 1000.
moab_df = df[
    (df['City'] == 'Moab') &
    (df['Year'] >= 2000) &
    ((df['New Residental Valuation'] + df['Value of Residental Additions']) > 1000)
]
```

```python
moab_df
```

```python
# Find the length of the Moab query
len(moab_df)
```
